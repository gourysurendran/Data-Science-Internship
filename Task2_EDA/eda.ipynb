```python
# Task 2: Exploratory Data Analysis (EDA)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("../Task1_Data_Pipeline/data.csv")

# 1. Survival Count
sns.countplot(x='Survived', data=data)
plt.title("Survival Count")
plt.show()

# 2. Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival by Gender")
plt.show()

# 3. Age Distribution
sns.histplot(data['Age'], bins=30)
plt.title("Age Distribution")
plt.show()

# 4. Fare Distribution
sns.histplot(data['Fare'], bins=30)
plt.title("Fare Distribution")
plt.show()

# 5. Correlation Heatmap
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Key Insights

print("1. Most passengers did not survive.")
print("2. Females had higher survival rate than males.")
print("3. Most passengers were aged between 20–30.")
print("4. Fare distribution is right-skewed (many low fares, few high).")
print("5. Higher class passengers had better survival chances.")
```
