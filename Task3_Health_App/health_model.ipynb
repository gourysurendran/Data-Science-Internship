```python
import pandas as pd

# Create simple health dataset
data = {
    "Age": [25, 45, 35, 50, 23, 60, 40, 30],
    "BMI": [22, 28, 26, 31, 21, 33, 29, 24],
    "BP": [120, 140, 130, 150, 115, 160, 135, 125],
    "Risk": [0, 2, 1, 2, 0, 2, 1, 0]  # 0=Low,1=Medium,2=High
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Features and target
X = df[["Age", "BMI", "BP"]]
y = df["Risk"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained")
```

```python
import Pickle
with open("model.Pki","wb")as f:
        pickle.dump(model,f)
print("Model saved")
```
